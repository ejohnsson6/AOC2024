```python
from pathlib import Path
from utils.get_input import get_input
import numpy as np

# inp = Path("./test_input").read_text()
inp = get_input(6)
inp
```

```python
# %%timeit
map = np.array([list(row) for row in inp.split("\n") if row])
map
```

```python
def start_direction(map):
    if (map == "^").any():
        return "UP"
    elif (map == "v").any():
        return "DOWN"
    elif (map == ">").any():
        return "RIGHT"
    elif (map == "<").any():
        return "LEFT"


def turn_90_degrees(dir):
    if dir == "UP":
        return "RIGHT"
    elif dir == "RIGHT":
        return "DOWN"
    elif dir == "DOWN":
        return "LEFT"
    elif dir == "LEFT":
        return "UP"


def start_position(map):
    return np.where(map == "^")


def is_out_of_bounds(y, x, max_y, max_x):
    return y < 0 or y >= max_y or x < 0 or x >= max_x


def part1():
    current_direction = "UP"

    y, x = start_position(map)
    max_y, max_x = map.shape

    have_been_map = np.zeros(map.shape)

    while not is_out_of_bounds(y, x, max_y, max_x):
        have_been_map[y, x] = 1
        new_x, new_y = x.copy(), y.copy()
        if current_direction == "UP":
            new_y -= 1
        elif current_direction == "DOWN":
            new_y += 1
        elif current_direction == "LEFT":
            new_x -= 1
        elif current_direction == "RIGHT":
            new_x += 1
        if (
            not is_out_of_bounds(new_y, new_x, max_y, max_x)
            and map[new_y, new_x] == "#"
        ):
            current_direction = turn_90_degrees(current_direction)
        else:
            x, y = new_x, new_y

    return int(have_been_map.sum())



part1()
```

```python
%%timeit
part1()
```

```python
from tqdm import tqdm


def steps_to_out_of_bounds(y, x, current_direction, map, max_y, max_x):
    y = y.copy()
    x = x.copy()
    cur_dir = current_direction
    # num_steps = 0
    have_been_map = np.zeros(map.shape)
    while not is_out_of_bounds(y, x, max_y, max_x):
        have_been_map[y, x] += 1
        new_x, new_y = x.copy(), y.copy()
        if cur_dir == "UP":
            new_y -= 1
        elif cur_dir == "DOWN":
            new_y += 1
        elif cur_dir == "LEFT":
            new_x -= 1
        elif cur_dir == "RIGHT":
            new_x += 1
        if not is_out_of_bounds(new_y, new_x, max_y, max_x) and (
            map[new_y, new_x] == "#" or map[new_y, new_x] == "O"
        ):
            cur_dir = turn_90_degrees(cur_dir)
        else:
            x, y = new_x, new_y
            if (
                not is_out_of_bounds(new_y, new_x, max_y, max_x)
                and have_been_map[new_y, new_x] > 10
            ):
                return True
        # num_steps += 1
    return False


def part2():
    sum = 0

    max_y, max_x = map.shape
    start_y, start_x = start_position(map)
    for xi in tqdm(range(max_x)):
        for yi in range(max_y):
            if not (xi == start_x and yi == start_y):
                map_added_obs = map.copy()
                map_added_obs[yi, xi] = "O"
                n = steps_to_out_of_bounds(
                    start_y, start_x, "UP", map_added_obs, max_y, max_x
                )
                if n:
                    sum += 1

    return sum



part2()
```
